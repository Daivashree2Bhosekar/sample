## String Slicing  
-    0 1 2 3 4
-    I N D E X
-    -5-4-3-2-1

 - [start:end:step/condition]
 - It traverses from left to right 

```python
word = " Today is saturday"   # Print Sat from the string
print(word[1:6:2])
```

```python
print(word[-6:-4])
```

```python
word1 = "python" #print yth with negative index
print(word1[-5:-2])
```

```python
print(word1[::-1])   # reverse the string
```

### String Formatting
   - f-string
   - .format()
   - optional(% legacy Operator)
   - {}= Placeholder 

```python
print("add 2 numbers : ",2 + 4)
print(f"The addition of num {2} and num {4} is {2+4}")
```

```python
print("Product of 2 numbers is {}".format(7*7))
print("Today is 3rd {0} and year is {1}".format("jan", 2026))
```

```python
print("price of laptop is : %d " % 240000)   # %d = integer / %s string
```

### Reading (r) , writing (w) , appending (a) a file

```python
with open("notes.txt", "w") as f:
    a = f.write("This is our first file")
print("Successfully created file")
```

```python
with open("notes.txt","r")as f:
    a = f.read()
    print(a)
```

```python
with open("notes.txt","a") as f:
    a = f.write("this is new file")
    print("created")
    f.close
```
